```python
sns.histplot(df['density'], bins=30, kde=True)
plt.title("density Distribution")
plt.show()
#ploting histogram with the value of density 
```

```python
sns.boxplot(x=df['pH'])
plt.title("pH Outliers")
plt.show()
#ploting a boxplot with the value of pH(acidity)
```

```python
sns.pairplot(df[['alcohol', 'pH', 'quality', 'sulphates']], hue='alcohol')
plt.show()
#plotting a pair plot with data alcohol ,ph,quality,sulphate and comparing them accordingly
```

```python
sns.boxplot(x='citric acid', y='residual sugar', data=df)
plt.title("citric acid Distribution by residual sugar")
plt.show()
#plotting a boxplot with citric acid in x axis and residual sugar in y axis 
```

```python
sns.barplot(x='chlorides', y='density', data=df)
plt.title("density Rate by chlorides")
plt.xticks([0, 1], ['high', 'low'])
plt.show()
#plotting barplot with clorides rate on x axis and density on y axis and comparing wheather chloride and density ratio is high or low
```
